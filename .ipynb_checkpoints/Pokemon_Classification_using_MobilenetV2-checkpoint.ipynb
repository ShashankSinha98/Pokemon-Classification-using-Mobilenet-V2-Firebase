```python
folder_path = 'drive/My Drive/Big Pokedex'
```

```python
import os
folders = os.listdir(folder_path)
print(folders)
```

```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
label_dict = {
    "Aerodactyl":0,  
    "Bulbasaur":1,  
    "Charmander":2, 
    "Dratini":3,  
    "Fearow":4,  
    "Meowth":5,  
    "Pikachu":6,  
    "Psyduck":7,  
    "Spearow":8,  
    "Squirtle":9
}
```

```python
train_image_data = []
train_labels = []

train_path = os.path.join(folder_path,"train")
pokemons = os.listdir(train_path)

for pokemon in pokemons:
    pok_img_path = os.path.join(train_path,pokemon)
    images = os.listdir(pok_img_path)
    
    for img_name in images:
        img_path = os.path.join(pok_img_path,img_name)
        img = image.load_img(img_path,target_size=(224,224))
        img_data = image.img_to_array(img)
        img_label = label_dict[pokemon]

        train_image_data.append(img_data)
        train_labels.append(img_label)
  


```

```python
print("Images:",len(train_image_data))
print("Labels:",len(train_labels))
```

```python
import random
combined = list(zip(train_image_data,train_labels))
random.shuffle(combined)

train_image_data[:], train_labels[:] = zip(*combined)
```

```python
import numpy as np
XTrain = np.array(train_image_data)
YTrain = np.array(train_labels)

print(XTrain.shape,YTrain.shape)
```

```python
from keras.utils import np_utils
YTrain = np_utils.to_categorical(YTrain)
print(XTrain.shape,YTrain.shape)
```

```python
import tensorflow as tf
from tensorflow.python.keras.optimizers import Adam
from tensorflow.python.keras.layers import *
from tensorflow.python.keras.models import Model
import matplotlib.pyplot as plt
```

```python
mobile_model = tf.keras.applications.MobileNetV2(include_top=False,weights='imagenet',input_shape=(224,224,3))
mobile_model.summary()
```

```python
av1 = GlobalAveragePooling2D()(mobile_model.output)
fc1 = Dense(256,activation='relu')(av1)
d1 = Dropout(0.5)(fc1)
fc2 = Dense(10,activation='softmax')(d1)

model_new = Model(mobile_model.input,fc2)
model_new.summary()
```

```python
adam = tf.keras.optimizers.Adam(lr=0.00003)
model_new.compile(loss='categorical_crossentropy',optimizer=adam,metrics=['accuracy'])
```

```python
for ix in range(len(model_new.layers)):
  print(ix,model_new.layers[ix])
```

```python
for ix in range(72):
  model_new.layers[ix].trainable = False

model_new.compile(loss='categorical_crossentropy',optimizer=adam,metrics=['accuracy'])
model_new.summary()
```

```python
hist = model_new.fit(XTrain,YTrain,shuffle=True,batch_size=16,epochs=50,validation_split=0.2)
```

```python
h = hist.history

# Visualizing loss
plt.plot(h['loss'],'r',label='Loss')
plt.plot(h['val_loss'],'b',label='Val Loss')
plt.legend()
plt.show()
```

```python
# Visualizing accuracy
plt.plot(h['accuracy'],'r',label='Acc')
plt.plot(h['val_accuracy'],'b',label='Val Acc')
plt.legend()
plt.show()
```

```python
model_new.evaluate(XTrain,YTrain)
```

```python
keras_file = "Pokemon MobilenetV2 Model"
model_new.save(keras_file)
```

```python
converter = tf.lite.TFLiteConverter.from_saved_model(keras_file)
```

```python
tflite_model = converter.convert()
```

```python
open('pokemon_mobilenetv2.tflite','wb').write(tflite_model)
```
